```python
#code for the hangman game with hints
import random

# Word list with hints
words_with_hints = [
    ("python", "A popular programming language"),
    ("giraffe", "Tallest land animal"),
    ("astronaut", "Travels in space"),
    ("volcano", "Mountain that erupts"),
    ("hangman", "You're playing it right now!")
]

# Hangman stages (visual progress)
hangman_stages = [
    """
     ------
     |    |
     |    
     |   
     |   
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   
     |   
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |   
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |   
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / 
     |   
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |   
    _|_
    """
]

def play_hangman():
    word, hint = random.choice(words_with_hints)
    guessed = set()
    attempts = 6
    print("ğŸ§  Hint:", hint)

    while attempts >= 0:
        display_word = ''.join([letter if letter in guessed else '_' for letter in word])
        print(hangman_stages[6 - attempts])
        print("ğŸ”¤ Word:", ' '.join(display_word))
        print("â¤ï¸ Attempts left:", attempts)

        if display_word == word:
            print("ğŸ‰ You guessed it! The word was:", word)
            break

        guess = input("ğŸ‘‰ Enter a letter: ").lower()
        if guess in guessed:
            print("âš ï¸ You already guessed that letter.")
        elif guess in word:
            guessed.add(guess)
            print("âœ… Correct!")
        else:
            guessed.add(guess)
            attempts -= 1
            print("âŒ Wrong!")

        if attempts < 0:
            print(hangman_stages[-1])
            print("ğŸ’€ Game Over! The word was:", word)

# Run the game
while True:
    play_hangman()
    again = input("ğŸ” Play again? (y/n): ").lower()
    if again != 'y':
        print("ğŸ‘‹ Thanks for playing!")
        break
```

```python
#web scraper to fetch headlines from BBC News
import requests
from bs4 import BeautifulSoup

url="https://www.bbc.com/news"
f=open("urls.txt","w")
print("Welcome to the Web Scraper App!")
response =requests.get(url)
headlines =[]
line=BeautifulSoup(response.text,'html.parser')
for item in line.find_all('h2'):
    headlines.append(item.get_text())       
for headline in headlines:
    f.write(headline + "\n")
print("Headlines have been saved to 'urls.txt'.")
f.close()
print("Exiting the Web Scraper App. Goodbye!")
```
